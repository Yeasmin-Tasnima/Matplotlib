```python
import matplotlib.pyplot as plt
%matplotlib inline
```

<h1>Create Data</h1>

```python
expanses = [15000, 3000, 1800, 20000, 2500]
chores = ["Home Rent", "Food", "Internet Bill", "Car", "Utilities"]
```

<h1>Pie Chart</h1>

```python
plt.axis('equal')
plt.pie(expanses, labels=chores)
plt.show()
```

<h3>Customize using Radius and Percentage (autopct)</h3>

```python
plt.axis('equal')
plt.pie(expanses, labels=chores, radius=1.5, autopct='%0.2f%%')
plt.show()
```

<h3>Customize using Explode (to bring out a pie)</h3>

```python
plt.axis('equal')
plt.pie(expanses, labels=chores, radius=1.5, autopct='%0.2f%%', explode=[0, 0.2, 0.5, 0, 0])
plt.show()
```

<h3>Customize using Start Angle</h3>

```python
plt.axis('equal')
plt.pie(expanses, labels=chores, radius=1.5, autopct='%0.2f%%', startangle=45)
plt.show()
```

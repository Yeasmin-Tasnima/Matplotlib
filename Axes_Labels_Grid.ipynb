```python
import matplotlib.pyplot as plt
%matplotlib inline
```

<h1>Create Data</h1>

```python
days = [1, 2, 3, 4, 5]
max_t = [100, 300, 200, 400, 500]
min_t = [50, 150, 120, 40, 200]
avg_t = [80, 200, 180, 300, 480]
```

<h1>Axes Labels, Legend</h1>

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')

plt.legend()
```

<h1>Grid</h1>

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')

plt.legend()
plt.grid()
```

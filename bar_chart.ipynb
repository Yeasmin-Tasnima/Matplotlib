```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

<h1>Create Data</h1>

```python
company = ['Google', 'Microsoft', 'Amazon', 'Facebook']
revenue = [300, 500, 200, 400]
profit = [30, 50, 20, 40]
```

<h3>Create numpy array corresponding to Company</h3>

```python
company_value = np.arange(len(company))
company_value
```

<h3>Bar Plot</h3>

```python
plt.bar(company_value, revenue)
```

<h3>Customize x plots and others</h3>

```python
plt.xticks(company_value, company)

plt.xlabel('company')
plt.ylabel('revenue')
plt.title('Stock Revenue')

plt.bar(company_value, revenue)
```

<h1>Bar plot of two variables</h1>

```python
plt.xticks(company_value, company)

plt.xlabel('company')
plt.ylabel('revenue')
plt.title('Stock Revenue')

plt.bar(company_value, revenue, label='Revenue')
plt.bar(company_value, profit, label='Profit')
plt.legend()
```

<h3>Change Position</h3>

```python
plt.xticks(company_value, company)

plt.xlabel('company')
plt.ylabel('revenue')
plt.title('Stock Revenue')

plt.bar(company_value - 0.1, revenue, width=0.5, label='Revenue')
plt.bar(company_value, profit, width=0.5, label='Profit')
plt.legend()
```

<h3>According to Y position (barh)</h3>

```python
plt.yticks(company_value, company)

plt.xlabel('company')
plt.ylabel('revenue')
plt.title('Stock Revenue')

plt.barh(company_value - 0.1, revenue, label='Revenue')
plt.barh(company_value, profit, label='Profit')
plt.legend()
```
